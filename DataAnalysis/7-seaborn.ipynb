```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Tips dataset")
plt.show()

```

#### lineplot

```python
sns.lineplot(data=tips, x="size", y="tip")
plt.title("Tips dataset")
plt.show()
```

```python
sns.barplot(data=tips, x="size", y="tip")
plt.title("Tips dataset")
plt.show()
```

#### BOX PLOT

```python
sns.boxplot(data=tips, x="size", y="tip")
```

```python
sns.violinplot(data=tips, x="size", y="tip")
plt.title("Tips dataset")
```

```python
sns.histplot(data=tips, x="tip", bins=10, kde=True)
```

```python
sns.kdeplot(data=tips, x="tip", fill=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate example data (customer purchase amounts)
np.random.seed(42)
data = np.concatenate([np.random.normal(50, 10, 500), np.random.normal(100, 20, 300)])

# Without KDE (Histogram)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color='lightblue', edgecolor='black', alpha=0.7)
plt.title('Histogram (Without KDE)')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')

# With KDE
plt.subplot(1, 2, 2)
kde = gaussian_kde(data)
x = np.linspace(data.min() - 10, data.max() + 10, 1000)
plt.plot(x, kde(x), color='darkblue', label='KDE')
plt.fill_between(x, kde(x), color='lightblue', alpha=0.5)
plt.title('Kernel Density Estimation (With KDE)')
plt.xlabel('Purchase Amount')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```
