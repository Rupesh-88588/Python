### Data Mainpulation and analysis with pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')

# fetch 5 rows
df.head(5)
```

```python
## Handling missing data

df.isnull().any(axis=1)
```

```python
df.isnull().sum()
```

```python
data_copied=df.fillna(0)
data_copied.head(5)
```

```python
## renaming columns
df=df.rename(columns={'Date':'Sales Date'}) 
df.head(5)
```

```python
## Change datatypes
df['Values_of_sales']=df['Value'].fillna(df['Value'].mean()).astype(float)
df.head(5)
```

```python
df['New_Value']=df['Value'].apply(lambda x: x*2)
df.head()
```

```python
### Data Aggregting AND Grouping

grouped_mean=df.head().groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_mean=df.groupby(['Product','Region'])['Value'].mean()
print(grouped_mean)
```

```python
## Aggegating with multiple functions
grouped=df.groupby(['Product','Region'])['Value'].agg(['sum','min','max'])
print(grouped)
```

```python
## mergging and joing dataframes
df1 = pd.DataFrame({'Keys': ['A', 'B', 'C', 'D'],'Values1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'Keys': ['A', 'B', 'E', 'F'],'Values2': [5, 6, 7, 8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Keys',how='inner') 
```

```python
pd.merge(df1,df2,on='Keys',how='left') 
```

```python
pd.merge(df1,df2,on='Keys',how='right') 
```
