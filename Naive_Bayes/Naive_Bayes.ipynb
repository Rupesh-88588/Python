```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print('accuracy score is :: ',accuracy_score(y_test,y_pred))
print('confusion matrix is :: ',confusion_matrix(y_test,y_pred))
print('classification report is :: ',classification_report(y_test,y_pred))

```

```python
from sklearn.model_selection import train_test_split

X = df[['total_bill', 'tip', 'sex', 'smoker','day', 'size']]
y = df['time']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.columns
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le1=LabelEncoder()
le2=LabelEncoder()
X_train['smoker']=le1.fit_transform(X_train['smoker'])
X_train['sex']=le2.fit_transform(X_train['sex'])
```

```python
X_test['smoker']=le1.transform(X_test['smoker'])
X_test['sex']=le2.transform(X_test['sex'])
```

```python
X_train
```

```python
X_train.columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(drop='first'),[4])],remainder='passthrough')

X_train=ct.fit_transform(X_train)
X_test=ct.transform(X_test)



```

```python
X_train
```

```python
import sys
import numpy as np

np.set_printoptions(threshold=sys.maxsize)
```

```python
X_train
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)  # Fit and transform training labels
y_test = le.transform(y_test) 
```

```python
from sklearn.naive_bayes import GaussianNB

gnb=GaussianNB()

gnb.fit(X_train,y_train)

```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print('accuracy score is :: ',accuracy_score(y_test,y_pred))
print('confusion matrix is :: ',confusion_matrix(y_test,y_pred))
print('classification report is :: ',classification_report(y_test,y_pred))
```
